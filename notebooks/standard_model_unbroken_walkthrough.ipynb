# Non-BFM Unbroken Standard Model

This notebook builds the unbroken Standard Model from `src/model/standard_model_unbroken.py`, reconstructs the named Lagrangian sections, and evaluates a few representative vertices.

```python
import sys
from pathlib import Path

ROOT = Path.cwd().resolve()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent
SRC = ROOT / "src"
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))
if str(SRC) not in sys.path:
    sys.path.insert(0, str(SRC))

from model import build_unbroken_standard_model
```

```python
def show(title, result):
  """Print vertex signatures and rules in a notebook-friendly format."""
  print(title)
  if isinstance(result, dict):
    print(f"{len(result)} vertex signature(s)\n")
    for signature, expression in result.items():
      print("Vertex:", signature)
      print("Rule:", expression)
      print()
  else:
    print(result)
    print()
def show_model(model, *fields, compact_form=None, sum_notation=None):
    source_terms = model.lagrangian_decl.source_terms
    if source_terms:
        lagrangian_source = sum(source_terms[1:], source_terms[0]) if len(source_terms) > 1 else source_terms[0]
        show("Lagrangian", lagrangian_source)
    lagrangian = model.lagrangian()
    if fields:
        show("Feynman Rule", lagrangian.feynman_rule(*fields, include_delta=False))
    else:
        show("Feynman Rules", lagrangian.feynman_rules(include_delta=False))
    if compact_form is not None:
        show("Compact Form", compact_form)
    if sum_notation is not None:
        show("Sum Notation", sum_notation)
```

```python
sm = build_unbroken_standard_model()

LGauge = sm.lagrangians.LGauge
LFermions = sm.lagrangians.LFermions
LHiggs = sm.lagrangians.LHiggs
LYukawa = sm.lagrangians.LYukawa
LSM = LGauge + LFermions + LHiggs + LYukawa

model = sm.model
L = model.lagrangian()
```

```python
{
    "validate_ok": model.validate().ok,
    "source_terms": len(LSM.source_terms),
    "compiled_terms": len(L.terms),
}
```

```python
interesting_vertices = {
    "qLbar qL G": L.feynman_rule(
        sm.fields.qL.bar,
        sm.fields.qL,
        sm.fields.G,
        simplify=True,
        include_delta=True,
    ),
    "Phi.bar Phi Wi B": L.feynman_rule(
        sm.fields.Phi.bar,
        sm.fields.Phi,
        sm.fields.Wi,
        sm.fields.B,
        simplify=True,
        include_delta=True,
    ),
    "qLbar Phi.bar uR": L.feynman_rule(
        sm.fields.qL.bar,
        sm.fields.Phi.bar,
        sm.fields.uR,
        simplify=True,
        include_delta=True,
    ),
}

interesting_vertices
```

```python
show_model(model)
```
