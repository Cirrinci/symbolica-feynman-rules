## Kinetic term --> propagator

The two point function is defined by the legs p1 and p2. we have the final result of 

K  = -I* m^2+I^3* p1* p2= -i(p1p2 + m^2)

convention both incoming therefore: p1=-p2 so we get the correct inverse propagator!


````python
# Kinetic + mass Lagrangian for a real scalar and its 2-point kernel
#
#   L = 1/2 (∂_μ φ)(∂^μ φ) - 1/2 m^2 φ^2



# Kinetic part: 1/2 (∂_μ φ)(∂^μ φ)
L_kin = LagrangianTerm(
    "kin",
    0.5 * delx(Phi(), mu) * delx(Phi(), mu),
    fields=("phi", "phi"),
    coefficient=0.5,
    factors=[
        OperatorFactor(Phi, derivative_indices=(mu,)),
        OperatorFactor(Phi, derivative_indices=(mu,)),
    ],
)

# Mass part: already defined in examples_scalar as L_scalar_mass = -1/2 m^2 φ^2
L_mass_nb = L_scalar_mass

legs_2pt = [
    ExternalLeg(Phi, "p1"),
    ExternalLeg(Phi, "p2"),
]

print("=== 2-point kernels from kinetic and mass terms ===")
k_kin  = canonical_vertex(L_kin, legs_2pt)
k_mass = canonical_vertex(L_mass_nb, legs_2pt)
print("kinetic kernel =", k_kin)
print("mass kernel    =", k_mass)
print("total kernel   =", (k_kin + k_mass))
print("total kernel (expanded) =", (k_kin + k_mass).expand())

from model import fast_bosonic_vertex
print("\n=== fast_bosonic_vertex check ===")
k_kin_fast  = fast_bosonic_vertex(L_kin, legs_2pt)
k_mass_fast = fast_bosonic_vertex(L_mass_nb, legs_2pt)
print("kinetic kernel (fast) =", k_kin_fast)
print("mass kernel (fast)    =", k_mass_fast)
print("total kernel (fast)   =", (k_kin_fast + k_mass_fast).expand())
````
